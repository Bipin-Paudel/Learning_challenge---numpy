# Advance operation with Business Example

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data structure: [resturant_id, 2021, 2022, 2023, 2024  ]

sales_data = np.array([
    [1,150000, 180000, 220000, 225000],  #Biryani house
    [2,150000, 190000, 200000, 105000],  #Momo  hotel
    [3,150000, 195000, 250000, 155000],  #Chowmin hotel
    [4,150000, 160000, 290000, 275000],  #Pizza hub
    [5,150000, 140000, 270000, 245000],  #chai point
])

print("== foodmandu sales analysis ===")

print("\n sample data for first 3 resturant  ", sales_data[0:3])
print("\n sales data shape  ", sales_data.shape)
```

```python
# total sales per year 

print(np.sum(sales_data, axis=0))
yearly_total = np.sum(sales_data[: , 1:], axis=0 )

print(yearly_total)
```

```python
# minimum sales per year 

min_sales = np.min(sales_data[: , 1:], axis=1 )
print(min_sales)

# maximum sales per year 

max_sales = np.max(sales_data[: , 1:], axis=0)
print("\n maximum sales in year ", max_sales)

# average sales per resturant
avg_sales = np.mean(sales_data[: , 1:], axis=0)
print("\n average sales in year ", avg_sales)


cumsum = np.cumsum(sales_data[: , 1:], axis=0)
print("cumulative sum of resturant is  ", cumsum)


plt.Figure(figsize=(9,6))
plt.plot(np.mean(cumsum, axis=0))
plt.title(" Average cumulative sales across all resturant ")
plt.xlabel("years")
plt.ylabel("sales")
plt.grid(True)
plt.show()
```

```python

vector_1 = np.array([1,2,3,4,5])
vector_2 = np.array([6,7,8,9,10])

print("\n vector matrix addition ", vector_1 + vector_2)
print("\n vector matrix multiplication  ", vector_1*vector_2)
print("\n Dot product ", np.dot(vector_1, vector_2))

angle = np.arccos(np.dot(vector_1, vector_2)/ (np.linalg.norm(vector_1)*np.linalg.norm(vector_2)))
print(angle)
```
