```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
array1= np.array([1,2,3,4,5])
array2 = np.random.rand(2,3)
array3 = np.zeros((3,4))

print("\n first array is ", array1)
print("\n second array is ", array2)
print("\n third array is ", array3)
```

```python
np.save("array1.npy", array1)
np.save("array2.npy", array2)
np.save("array3.npy", array3)
```

```python
loaded_array1 = np.load('array1.npy')
print(loaded_array1)
```

```python
try:
    logo = np.load("numpy-logo.npy")
 # Display

    plt.figure(figsize=(10,5))
    plt.subplot(121)
    plt.imshow(logo)
    plt.title("Numpy logo")
    plt.grid(False)

    # dark logo
    dark_logo = 1- logo
    plt.figure(figsize=(10,5))
    plt.subplot(122)
    plt.imshow(dark_logo)
    plt.title("Numpy logo dark")
    plt.grid(False)



except FileNotFoundError:
    print("numpy logo file not found")
```
